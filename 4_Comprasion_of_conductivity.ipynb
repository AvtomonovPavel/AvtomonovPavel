```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_1 = 'https://raw.githubusercontent.com/AvtomonovPavel/Method-of-sources/main/Examples/4.%20Fracture%20-%20Uniform%20flux.txt'
path_2 = 'https://raw.githubusercontent.com/AvtomonovPavel/Method-of-sources/main/Examples/4.2%20Fracture%20-%20infinity%20conductivity.txt'
df_uniform = pd.read_table(path_1, sep='\s+', engine = 'python')
df_inf = pd.read_table(path_2, sep='\s+', engine = 'python')
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(16, 8)

ax1.plot(df_uniform['dTime'], df_uniform['p-p@dt=0'],'x', label = 'Депрессия. Uniform flux.')
ax1.plot(df_uniform['dTime'], df_uniform['Derivative'],'x', label = 'Производная депрессии. Uniform flux.')
ax1.plot(df_inf['dTime'], df_inf['p-p@dt=0'],'o', label = 'Депрессия. Infinity conductivity.')
ax1.plot(df_inf['dTime'], df_inf['Derivative'],'o', label = 'Производная депрессии. Infinity conductivity')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.legend()
plt.title("Трещина гидроразрыва пласта. dP(t)")
ax1.set_xscale('log')
ax1.set_yscale('log')
plt.xlabel("t, c")
plt.ylabel("dP, Па")
ax1.grid()
ax1.legend()
```
